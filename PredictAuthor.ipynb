```python
import re
import json
import sys
import collections
import os
import json
from datetime import datetime
startTime = datetime.now()


bigrams = collections.defaultdict(lambda: 0)
sentences = []

book_title = 'GreatExpectations.txt'

if os.path.isfile(book_title):
    print('Processing file', book_title)
    with open(book_title, 'r') as f0:
        sentence_buffer = []
        for i, line in enumerate(f0.readlines()):
            if i % 100 == 0:
                print('Processed', i, 'lines.')
            line = line.rstrip()
            if len(line) < 1:
                continue
            #if re.search(r'^[A-Z][A-Z][A-Z]', line):
            #    continue
            if line[0] == '[':
                continue
            line = re.sub(r'([\.,;:!\?”])', r' \1', line)
            line = re.sub(r'(“)', r'\1 ', line)
            line = re.sub(r'[_‘]', '', line)
            line = re.sub('—', ' ', line)
            #line = re.sub(r'[^a-zA-Z\.’ ]', '', line)
            #print(line)
            buffer_empty = True
            lal = line.split()
            for i, wd in enumerate(lal):
                if i > 0 and buffer_empty == False:
                    bigrams[lal[i-1]+'_'+wd] += 1
                #wd = wd.lower()
                buffer_empty = False
                if wd in ['.', '!', '?', ':', ';']:
                    buffer_empty = True
else:
    print('No file found with  name', book_title)
    exit()
    
total_bigrams = len(bigrams)
print('There are', total_bigrams, 'bigrams in', book_title)
for bigram in list(bigrams):
    print(bigram, bigrams[bigram])
    bigrams[bigram] = bigrams[bigram] / total_bigrams

with open(book_title+'_bigrams.json', 'w') as f1:
    json.dump(bigrams, f1)
    print('json file created')
```

```python
import re
import json
import sys
import collections
import os
import json
import math
from datetime import datetime
startTime = datetime.now()

book_title1 = 'emma.txt'
book_title2 = 'tale_of_2_cities.txt'
bigrams1 = 'pride_and_prejudice.txt_bigrams.json'
bigrams2 = 'GreatExpectations.txt_bigrams.json'

with open(bigrams1, 'r') as f1:
    bigram_dict1 = json.load(f1)

with open(bigrams2, 'r') as f2:
    bigram_dict2 = json.load(f2)

bigrams_in_both_dictionaries = [b for b in list(bigram_dict1) if b in list(bigram_dict2)]

for test_file in [book_title1, book_title2]:
    bigrams_tested1 = 1
    bigrams_tested2 = 1
    log_prob1 = 0
    log_prob2 = 0
    print('Processing file', test_file)
    with open(test_file, 'r') as f0:
        sentence_buffer = []
        for i, line in enumerate(f0.readlines()):
            if i % 1000 == 0:
                print('Processed', i, 'lines with', bigrams_tested1, 'bigrams1 tested and', bigrams_tested2, 'bigrams2 tested')
            line = line.rstrip()
            if len(line) < 1:
                continue
            #if re.search(r'^[A-Z][A-Z][A-Z]', line):
            #    continue
            if line[0] == '[':
                continue
            line = re.sub(r'([\.,;:!\?”])', r' \1', line)
            line = re.sub(r'(“)', r'\1 ', line)
            line = re.sub(r'[_‘]', '', line)
            line = re.sub('—', ' ', line)
            #line = re.sub(r'[^a-zA-Z\.’ ]', '', line)
            #print(line)
            buffer_empty = True
            lal = line.split()
            for i, wd in enumerate(lal):
                if i > 0 and buffer_empty == False:
                    bigram_to_test = lal[i-1]+'_'+wd
                    if bigram_to_test in bigrams_in_both_dictionaries: 
                    #if bigram_to_test in list(bigram_dict1):
                        log_prob1 += math.log(bigram_dict1[bigram_to_test])
                        bigrams_tested1 += 1
                    #if bigram_to_test in list(bigram_dict2):
                        log_prob2 += math.log(bigram_dict2[bigram_to_test])
                        bigrams_tested2 += 1
                buffer_empty = False
                if wd in ['.', '!', '?', ':', ';']:
                    buffer_empty = True
    print('Average log prob for file', test_file, 'on model 1 =', log_prob1 / bigrams_tested1) 
    print('Average log prob for file', test_file, 'on model 2 =', log_prob2 / bigrams_tested2) 
print(datetime.now() - startTime)
```
